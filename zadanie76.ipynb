```python
import numpy as np
import numpy_financial as npf

initial_price = 120_000  
growth_rate = 0.05       
years = 5
months = years * 12


price_after_5_years = initial_price * (1 + growth_rate) ** years
print(f"Cena mieszkania po 5 latach: {price_after_5_years:.2f} zł")
```

```python
nominal_interest_rate = 0.12
monthly_rate = nominal_interest_rate / 12
monthly_payment = -npf.pmt(rate=monthly_rate, nper=months, pv=0, fv=price_after_5_years)
print(f"Niezbędna miesięczna wpłata: {monthly_payment:.2f} zł")
```
