```python
import numpy as np
import numpy_financial as npf

initial_price = 120_000
growth_rate = 0.05
years = 5
months = years * 12


price_after_5_years = initial_price * (1 + growth_rate) ** years
print(f"Cena mieszkania po 5 latach: {price_after_5_years:.2f} zł")
```

```python
nominal_interest_rate = 0.12
monthly_rate = nominal_interest_rate / 12
monthly_payment = -npf.pmt(rate=monthly_rate, nper=months, pv=0, fv=price_after_5_years)
print(f"Niezbędna miesięczna wpłata: {monthly_payment:.2f} zł")
```

```python
from matplotlib import pyplot as plt

n = np.arange(1, months + 1)
deposit_value = monthly_payment * ((1 + monthly_rate) ** n - 1) / monthly_rate

apartment_price = np.linspace(initial_price, price_after_5_years, months)
months_axis = n

plt.figure(figsize=(10, 6))
plt.plot(months_axis, apartment_price, label="Cena mieszkania", linestyle="-")
plt.plot(months_axis, deposit_value, label="Wartość lokaty", linewidth=2)
plt.title("Wzrost ceny mieszkania vs. wartość lokaty")
plt.xlabel("Miesiąc")
plt.ylabel("Wartość [zł]")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
